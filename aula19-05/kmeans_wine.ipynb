```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
vinho = load_wine()
vinho.target_names
vinho.feature_names
```

```python
X = vinho.data
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='random')
    kmeans.fit(X)
    print(i, kmeans.inertia_)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Metodo Elbow')
plt.xlabel('Numero de clusters')
plt.ylabel('WCSS') # soma dos quadrados do cluster
plt.show()
```

```python
# KMEANS com 3 clusters
# aplicando algoritmo
kmeans = KMeans(n_clusters=3, init='random')
kmeans.fit(X)

# examinando dados agrupados (centroides)
kmeans.cluster_centers_

distance = kmeans.fit_transform(X)
distance

# conferindo dados do cluster
labels = kmeans.labels_
labels
```

```python
data = [[4.12, 3.4, 1.6, 0.7, 5.1, 3.0, 1.8, 2.5, 3.9, 0.9, 4.1, 1.1, 2.0]]
kmeans.predict(data)
```

```python
plt.scatter(X[:,0], X[:,1], s=100, c=kmeans.labels_)
plt.title('Vinhos clusters e centroides')
plt.xlabel('Alcohol')
plt.ylabel('Malic_acid')
plt.legend()
plt.show()
```
